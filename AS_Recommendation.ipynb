```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv("anime.csv")
df.head()
```

```python
df.info()
df.describe()
df.isnull().sum()
df = df.dropna()
```

```python
df = df[['name', 'genre', 'rating', 'members']]
df.head()
```

```python
genre_dummies = df['genre'].str.get_dummies(sep=',')
```

```python
features = pd.concat([genre_dummies, df[['rating', 'members']]], axis=1)
```

```python
features.head
```

```python
scaler = MinMaxScaler()
features[['rating', 'members']] = scaler.fit_transform(features[['rating', 'members']])
```

```python
features.head
```

```python
similarity_matrix = cosine_similarity(features)
similarity_df = pd.DataFrame(
    similarity_matrix,
    index=df['name'],
    columns=df['name']
)
similarity_df.head()
```

```python
def recommend_anime(anime_name, top_n=5):

    if anime_name not in similarity_df.index:
        return "Anime not found"

    scores = similarity_df[anime_name]

    scores = scores.sort_values(ascending=False)

    scores = scores[1:top_n+1]

    return scores
```

```python
recommend_anime("Naruto")
```

```python
def recommend_with_threshold(anime_name, threshold=0.5):

    if anime_name not in similarity_df.index:
        return "Anime not found"

    scores = similarity_df[anime_name]

    scores = scores[scores > threshold]

    scores = scores.sort_values(ascending=False)

    scores = scores[1:]

    return scores
```

```python
print("Recommendations with Threshold = 0.95")
recommend_with_threshold("Naruto", threshold=0.95)
```

```python
print("Recommendations with Threshold = 0.80")
recommend_with_threshold("Naruto", threshold=0.80)
```

```python
print("Recommendations with Threshold = 0.50")
recommend_with_threshold("Naruto", threshold=0.50)
```

```python

Observation

Threshold = 0.95
Very few recommendations
Highly similar anime only

Threshold = 0.80
Balanced recommendations
Good accuracy and diversity

Threshold = 0.50
More recommendations
Includes less similar anime

Conclusion
Threshold selection controls
recommendation quality and quantity.
```

```python
#Q1
User-Based Collaborative Filtering:
Recommends items based on similar users.
If User A and User B have similar interests,
items liked by User B are recommended to User A.

Example
User A likes Naruto and One Piece.
User B likes Naruto, One Piece, and Bleach.
System recommends Bleach to User A.

Advantages:
Personalized recommendations

Disadvantages:
Slow for large datasets


Item-Based Collaborative Filtering:
Recommends items based on item similarity.
If a user likes Naruto,
similar anime are recommended.

Example
Naruto is similar to Bleach and One Piece.
System recommends them

Advantages:
Faster and scalable
Stable recommendations

Disadvantages:
Needs item similarity computation

```

```python
#Q2

Collaborative Filtering is a recommendation technique
that suggests items based on user behavior and similarity.

Working Steps

1. Collect user-item interaction data
   Example:
   ratings, watch history, likes

2. Create feature matrix

3. Calculate similarity
   - User similarity or
   - Item similarity

4. Recommend items with highest similarity


Types:
1. User-Based Collaborative Filtering
2. Item-Based Collaborative Filtering

Advantages:
Personalized recommendations
No need for detailed item information

Disadvantages:
Cold start problem
Sparse data issue
Scalability challenges
```
